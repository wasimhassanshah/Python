```python
def max_count(lst):
    max_count = 0
    max_occured_num = None
    
    for num in range(len(lst)):
        count = lst.count(num) # how many times any unique number occurs in the list
        if count> max_count:
            max_count= count
            max_occured_num = num
    return max_occured_num
```

```python
my_list = [1, 2, 3, 4, 2, 2,2,2, 3, 3, 3, 5]
result=(max_count(my_list))

print("Most common number: ", result)
```

A hash map is a data structure that allows you to store and retrieve key-value pairs in an efficient way.

The enumerate function is used to iterate over a sequence (such as a list) while keeping track of the index of the current item. It returns pairs of the form (index, element).

In the context of the for i, num in enumerate(nums): line:

i is the index of the current element in the iteration. num is the actual element in the nums list.

```python
# Hash map approach


def two_sums(lst, target):
    num_dict = {}
    
    for i, num in enumerate(lst):
        complement= target-num
        if complement in num_dict:
            return [num_dict[complement],i] # return the indices of those two numbers as a list.

        num_dict[num]=i   #If the complement is not found in the dictionary, we update the num_dict by adding the
                        #current number as a key and its index as the corresponding value. 
                   # This is done to keep track of the numbers encountered so far.
```

```python
arr=[2,7,11,15]
target=17

result=two_sums(arr,target)

print(result)
```

```python
def three_sum(arr, target):
    for i in range(len(arr)):
        for j in range(i+1, len(arr)):
            for k in range(j+1, len(arr)):
                if arr[i]+arr[j]+arr[k]==target:
                    return [i,j,k]
    return None
               
            
```

```python
arr=[1,0,1,2,-1,4]
target=7

three_sum(arr,target)
```

```python
# Hash map

def three_sum_hash(arr:list, target):
    num_dict={}
    for i in range(len(arr)):
        for j in range(i+1,len(arr)):
            complement = target -(arr[i]+arr[j])
            if complement in num_dict:
                return [num_dict[complement],i,j]
            
            num_dict[arr[i]]=i
            num_dict[arr[j]]=j
    return None

```

The lines num_dict[arr[i]] = i and num_dict[arr[j]] = j are updating a dictionary (num_dict) by associating values from the input list (arr) with their corresponding indices in the list. 

```python
arr=[1,2,3,4,5,6]
target=12

three_sum_hash(arr,target)
```

```python
# Given a square matrix, calculate the absoulte difference between the sums of its diagonals

def diag_diff(arr):
    right_sum = 0
    left_sum = 0
    n=len(arr)
    
    for i in range(n):
        right_sum += arr[i][i]
        left_sum += arr[i][n-1-i]
        
    return abs(right_sum-left_sum)
    
    
```

```python
matrix=[
    [11,2,4],
    [4,5,6],
    [10,8,-12]
]

diag_diff(matrix)
```

# Transpose matrix

transposed_matrix = [[0] * m for _ in range(n)]

[[0] * m for _ in range(n)]: For each iteration, it creates a new list [0] * m, which represents a row in the matrix, filled with m zeros. This list comprehension is repeated n times, creating a list of lists.

```python
 transposed_matrix = [[0]*m for _ in range(n)]  # generally m is row num and n is column number but here m becomes 
                                            #column number and n row numbers

transposed_matrix = [[0]*3 for _ in range(2)]

transposed_matrix 
```

```python
def transpose_matrix(matrix):
    # Get num of rows and columns in original matrix
    
    m,n= len(matrix), len(matrix[0])
    
    # structure of transpose matrix would be n*m (which is transpose of m*n) (row*columns)
    
    transposed_matrix = [[0]*m for _ in range(n)]
    
    for i in range(m):
        for j in range(n):
            transposed_matrix[j][i]=matrix[i][j]
    return transposed_matrix
    
    
```

```python
matrix=[[1,2,3],[4,5,6],[7,8,9]]

transposed_matrix1 = transpose_matrix(matrix
                                     )

transposed_matrix1
```

```python
# Duplicates using Hash map


def if_dup(arr):
    hash_set= set()
    
    for n in arr:
        if n in hash_set:
            return True
        hash_set.add(n)
    return False
    
```

```python
n = [1,2,3,4,5,5]

if_dup(n)
```

The line max_diff=float('-inf') initializes a variable named max_diff with the value negative infinity.

In Python, float('-inf') represents negative infinity, which is a special floating-point value that is smaller than any finite number. It's often used as an initial value when you want to find the maximum (or minimum) value in a set of numbers.



```python
def special_diff(k):
    max_diff=float('-inf')
    n = len(k)
    
    for a in range(n):
        for b in range(a+1,n):
            for c in range(n):
                for d in range(c+1,n):
                    special_diff=(k[a]*k[b])-(k[c]*k[d])
                    max_diff = max(max_diff, special_diff)
    return max_diff
        
    
```

```python
k = [3,2,1,3,4,5,6]

result= special_diff(k)
result
```

```python
def max_sub(nums):
    max_sum = nums[0]
    curr_sum = 0
    for n in nums:
        if curr_sum<0:
            curr_sum=0
        curr_sum+=n
        if max_sum<curr_sum:     # it is similar as max_sum=max(max_sum,curr_sum)
            max_sum=curr_sum
    return max_sum
        
    
```

```python
nums=[-2,1,-3,4,-1,2,1,-5,4]

max_sub(nums)
```

```python
def left_rotate(arr,d):
    
    rotated_arr= arr[d:]+arr[:d] 
    
    return rotated_arr
```

```python
d=4
arr=[1,2,3,4,5]

result= left_rotate(arr,d)

print(result)
```

```python
def right_rotate(arr,d):
    
    rotated_arr_r= arr[-d:]+arr[:-d]
    
    return rotated_arr_r
```

```python
d=4
arr=[1,2,3,4,5]

result= right_rotate(arr,d)

print(result)
```

```python
def bubble_sort(arr):
    swap=0
    
    for i in range(len(arr)-1):
        for j in range(i+1, len(arr)):
            if arr[i]>arr[j]:
                arr[i],arr[j]=arr[j],arr[i]
                swap+=1
    
    print("sorted array ", arr)
    print("first element ", arr[0])
    print("last element ", arr[-1])
    print("swaps happened for sorting ",swap)
        
    
```

```python
arr=[6,4,1,7,0]

bubble_sort(arr)
```
