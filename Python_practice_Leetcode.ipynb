```python
def twoSum(num,target):
    prevMap = {} # Value: index
#enumerate . It's often used in for loops when you need to access both the elements and their indices within the iterable.
    for i, n in enumerate(num):
        diff= target - n  # n is element of list nums
        if diff in prevMap:  #checikgn if that difference value is in HashMap (prevMap) if yes return its index along 
             return [prevMap[diff],i]                  #with ith index
            
        prevMap[n]= i  # If not finding solution update hashmap
    return
```

```python
nums1 = [2, 7, 11, 15]
target1 = 9
result1 = twoSum(nums1, target1)
print(result1)  # Output: [0, 1]
```

```python
def containsDuplicate(nums):
    hashset= set()
    
    for n in nums:
        if n in hashset:
            return True
        hashset.add(n)
    return False


```

```python
nums=[1,2,3,4,5,6]

containsDuplicate(nums)
    
```

```python
nums2=[1,2,3,4,1,2,3,4]

containsDuplicate(nums2)
```

```python
def maxSubArray(nums:list[int]):
    maxSub= nums[0]
    curSum = 0
    for n in nums:
            if curSum<0:
                curSum = 0
            curSum += n
            maxSub= max(maxSub, curSum)
    return maxSub
            
```

```python
nums=[-2,1,-3,4,-1,2,1,-5,4]

maxSubArray(nums)
```

```python
def maxSubArray1(nums):
        max_sum = float('-inf')  # Initialize the maximum sum to negative infinity
        current_sum = 0  # Initialize the current sum to 0

        for num in nums:
            current_sum = max(num, current_sum + num)
            max_sum = max(max_sum, current_sum)

        return max_sum
```

```python
from collections import Counter

def sockMerchant(n, ar):
    # Count the frequency of each color in the array
    color_count = Counter(ar)
    
    # Initialize the total number of pairs
    total_pairs = 0
    
    # Calculate the number of pairs for each color
    for color, count in color_count.items():
        pairs = count // 2
        total_pairs += pairs
    
    return total_pairs

```

```python
# Example usage:
n = 9
ar = [10, 20, 20, 10, 10, 30, 50, 10, 20]
result = sockMerchant(n, ar)
print(result)  # Output: 3
```

```python
def countingValleys(steps, path):
    valley = 0
    current =0
    for i in range(steps):
        previous=current
        if path[i] == 'U':
            current+=1
        if path[i] == 'D':
            current-=1
        if current == 0 and i!=0:
            if previous==-1:
                valley+=1
    return valley
```

```python

path = "UDDDUDUU"

countingValleys(8, path)
```

```python
def repeatedString(s,n):   # s is tring  , n is number of characters repeating comprising string
    
    count_a_in_s= s.count('a') # suppose s= asa  it returns 2 as 2 a's inin string
    
    full_repeats = n//len(s)   #  it gives for n=10 and len(s)= 3 so full_repeats=10/3=9
                           # if n=10 means asaasaasaa # means 10 characters repeated string sequence wise
        
    total_a_in_repeats= count_a_in_s*full_repeats # it gives 7= 3 a's in one string * 2 full_repeats(number_of times string repeats)
    
    remaining_chars = n%len(s)  # it goves 10/3 = 1 word which is first word of string remaining to be count
    
    a_in_remaining = s[:remaining_chars].count('a') # Count a's in the remaining characters
    
    total_a_count = total_a_in_repeats + a_in_remaining
    
    return total_a_count
    
    
    
    
```

```python
s="asa"
n=10

result=repeatedString(s,n)

print(result)
```

```python
def rotateLeft(d,arr):
    n=len(arr)
    
    #Perfrom left rotation using array slicing
    
    rotated_arr = arr[d:] + arr[:d]
    
    return rotated_arr
```

```python
d=4
arr=[1,2,3,4,5]

result=rotateLeft(d,arr)

print(result)
```

```python
# Rotate array RIght

def rotateRight(d, arr):
    n=len(arr)
    
    rotate_arr= arr[-d:] + arr[:-d]
    
    return rotate_arr

```

```python
from collections import Counter

def checkMagazine(magazine, note):
    mag_counter = Counter(magazine)
    note_counter = Counter(note)
    
    result = note_counter - mag_counter  # Calculate the differences using Counters
    
    if not result:  # Check if the result is empty
        print("Yes")
    else:
        print("No")
```

```python
magazine1 = ["give", "me", "one", "grand", "today", "night"]
note1 = ["give", "one", "grand", "today"]

magazine2 = ["two", "times", "three", "is", "not", "four"]
note2 = ["two", "times", "two", "is", "four"]

magazine3 = ["ive", "got", "a", "lovely", "bunch", "of", "coconuts"]
note3 = ["ive", "got", "some", "coconuts"]

checkMagazine(magazine1, note1)  # Output: Yes
checkMagazine(magazine2, note2)  # Output: No
checkMagazine(magazine3, note3)  # Output: No
```

#Bubble Sort is a simple sorting algorithm that repeatedly steps through the list, compares adjacent elements, 
#and swaps them if they are in the wrong order. The pass through the list is repeated until no swaps are needed,
#indicating that the list is sorted. It is called "Bubble Sort" because the smaller (or larger, depending on the order)
#elements "bubble" to the top of the
#list in each pass.

```python
def countSwap(a):
    swap=0
    for i in range(len(a)-1):
        for j in range(i,len(a)):
            if a[i]>a[j]:
                a[i],a[j]=a[j],a[i]  #Swaping a[i] wi a[j] and a[j] with a[i] if a[i]>a[j]
                swap+=1
    print("Array is sorted in",swap,"swaps.")
    print("First element: ",a[0])
    print("Last element ",a[-1])
    print("Sorted array",a)
    return
```

```python
a=[6,4,1]

countSwap(a)
```

```python
def maximumToys(prices,k):
    # Sort the toy prices in ascending order
    prices.sort()
    
    toys_bought = 0
    budget = k
    
    
    for price in prices:
        if budget - price >= 0:
            budget -= price
            toys_bought += 1
        else:
            break
            
# budget - price >= 0:: This condition checks whether Mark's remaining budget after buying a toy (i.e., budget - price) 
#is greater than or equal to zero. If it's true, it means Mark can afford to buy the toy without going over his budget.

#budget -= price: If the condition is met (i.e., Mark can afford the toy), the toy's price is subtracted from Mark's budget.
#This reflects the purchase of the toy.

#toys_bought += 1: After successfully buying the toy, the variable toys_bought is incremented by 1. This variable keeps
#track of the number of toys Mark has bought.
            
    return toys_bought
        
        
    
```

```python
n=7
k=50

prices = [1,12,5,111,200,1000,10]

result = maximumToys(prices,k)

print(result)
```
