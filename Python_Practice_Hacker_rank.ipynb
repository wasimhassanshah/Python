# Factorial using loop

```python
def factorial(n):
    if n < 0:
        return "Factorial is not defined for negative numbers"
    elif n == 0 or n == 1:
        return 1
    else:
        result = 1
        for i in range(2, n + 1):
            result *= i
        return result

# Example usage:
n = 5
print(f"The factorial of {n} is {factorial(n)}")
```

# Check Perfect Numbers
A perfect number is a positive integer that is equal to the sum of its proper divisors, excluding itself. 
In other words, the sum of the positive divisors of a perfect number (excluding the number itself) equals the number itself.

6: The divisors of 6 are 1, 2, and 3. The sum of these divisors is 1 + 2 + 3 = 6, which is equal to the number itself.

28: The divisors of 28 are 1, 2, 4, 7, and 14. The sum of these divisors is 1 + 2 + 4 + 7 + 14 = 28.

```python
num= int(input("Enter a number :" ))

def perfect(num):
    sum=0
    for i in range(1, num):
        if num%i==0:
            sum=sum+i
    if sum==num:
         print("Entered number is Perfect number")
    else:
        print("Entered number is not a Perfect number")
        
        
perfect(num)
```

```python
def longest_palindrome_length(s):
    from collections import Counter

    # Create a dictionary to count the frequency of each character
    char_count = Counter(s)

    # Initialize variables to keep track of the length of the palindrome
    length = 0
    has_odd = False

    # Iterate through the character counts
    for count in char_count.values():
        # Add even counts to the length of the palindrome
        length += (count // 2) * 2

        # If there's an odd count, mark that the palindrome has an odd-length component
        if count % 2 == 1:
            has_odd = True

    # If there was at least one character with an odd count, add 1 to the length
    if has_odd:
        length += 1

    return length

# Example usage:
s = "abbccccdd"
result = longest_palindrome_length(s)
print(f"The length of the longest palindrome that can be built is {result}")
```

```python
def rem_zero(zero_end: list[int]):
    zero_count = zero_end.count(0)  # count number of zero in list

    for i in range(zero_count):
        zero_end.remove(0)  # remove 0 values from the list one by one and append them to the end for as many times as there are 0
        zero_end.append(0)

    return zero_end

```

```python
def basket_score(ops):
    scores=[]
    for op in ops:
        if op == '+':
            score=scores[-1]+scores[-2]
            scores.append(score)
        elif op == 'D':
            score=2*scores[-1]
            scores.append(score)
        elif op=='C':
            scores.pop()
        else:
            score=int(op)
            scores.append(score)
            
    return sum(scores)
```

```python
ops=[1,2,'C','D','+',5]

result=basket_score(ops)


# Print the result
print("Final score:", result)
```

```python

def isValid(s):
    stack = []
    mapping = {')': '(', ']': '[', '}': '{'}
    
    for char in s:
        if char in mapping:
            if not stack or stack[-1] != mapping[char]:
                return False
            stack.pop()
        else:
            stack.append(char)
    
    return not stack
```

```python
# Example usage
s = "([{}])"
print(isValid(s))  # Output: True

s = "()[]{}"
print(isValid(s))  # Output: True

s = "([)]"
print(isValid(s))  # Output: False

s = "{[}]"
print(isValid(s))  # Output: False
```

```python
def countPerfectBalls(k, x, y, z):
    count = 0
    
    for i in range(len(k)):
        for j in range(i + 1, len(k)):
            for n in range(j + 1, len(k)):
                if (
                    abs(k[i] - k[j]) <= x
                    and abs(k[j] - k[n]) <= y
                    and abs(k[i] - k[n]) <= z
                ):
                    count += 1
    
    return count
```

```python

# Example usage:
k = [3, 0, 1, 1, 9, 7]
x = 7
y = 2
z = 3

result = countPerfectBalls(k, x, y, z)

# Print the result
print("Final result:", result)
```

	Finally, we return the count, which represents the number of perfect ball sets.

```python
# Function to calculate the number of Ferus keystrokes
def ferus_keystrokes(message):
    vowels = 'aeiou'

    # Check if "avengers" is present in the message
    if "avengers" in message:
        # Exclude "avengers" from having its vowels removed
        parts = message.split("avengers")
        ferus_parts = []
        for part in parts:
            ferus_part = ''.join(char for char in part if char not in vowels)
            ferus_parts.append(ferus_part)
        ferus_message = "avengers".join(ferus_parts)
    else:
        # If "avengers" is not in the message, proceed as before
        ferus_message = ''.join(char for char in message if char not in vowels)

    return len(ferus_message)

# Function to calculate the number of normal phone keystrokes
def normal_phone_keystrokes(message):
    return len(message)

# Read the input message as a single string
message = input()

ferus_count = ferus_keystrokes(message)
normal_count = normal_phone_keystrokes(message)

print(f"{ferus_count}/{normal_count}")
```

```python
def three_sum_bruteforce(nums, target):
    n = len(nums)
    found = False

    for i in range(n):
        for j in range(i + 1, n):
            for k in range(j + 1, n):
                if nums[i] + nums[j] + nums[k] == target:
                    found = True

    return found

```

```python
# Example usage:
nums = [1, 0, 1, 2, -1, -4]
target = 5
result = three_sum_bruteforce(nums, target)
print(result)
```

```python
 def diagonalDifference(arr):
        n=len(arr)
        primary_sum = 0
        secondary_sum = 0
        
        for i in range(n):
            primary_sum += arr[i][i]
            secondary_sum += arr[i][n-1-i]
        
        return abs(primary_sum-secondary_sum)

#! In the code secondary_sum += arr[i][n - 1 - i], we are calculating the sum of elements along the secondary diagonal 
#of the square matrix

# n - 1 - i calculates the column index for the secondary diagonal. It's derived as follows:
# n - 1 represents the index of the last column in the matrix.
# Subtracting i from it gives us the column index that corresponds to the diagonal from the top right to the bottom left.
#This index decreases as we move from the top right to the bottom left along the secondary diagonal.




#Example


matrix= [
    [11,2,4],
    [4,5,6],
    [10,8,-12]
]



result=diagonalDifference(matrix)


print(result)
```

```python
import re

def is_valid_uid(uid):
    if len(uid) != 10:
        return "Invalid"
    
    if not re.match(r'^[A-Za-z0-9]*$',uid):
        return "Invalid"
    
    # re.match: This is a function from Python's re module, which is used for working with regular expressions.

    #r'^[A-Za-z0-9]*$': This is the regular expression pattern. Let's break it down:

    #^: Anchors the pattern at the start of the string.
    #[A-Za-z0-9]: This part defines a character class that matches any uppercase letter (A-Z), lowercase letter (a-z), or digit (0-9).
    #*: Matches zero or more occurrences of the preceding character class.
    #$: Anchors the pattern at the end of the string.
    
    if len(set(uid)) != len(uid):
        return "Invalid"
    
    #set(uid): The set function is used to create a set from the characters in the UID.
    #A set is an unordered collection of unique elements, which means it automatically removes any duplicate characters 
    #from the UID.

   #len(set(uid)): This part calculates the length (number of unique characters) of the set created from the UID.
    
    uppercase_count = sum(1 for char in uid if char.isupper())
    digit_count = sum(1 for char in uid if char.isdigit())
    
    if uppercase_count<2 or digit_count<3:
        return "Invalid"
    
    return "Valid"




# Read the number of test case

n= int(input("Enter the number of test cases: "))



#Process Each Test case

for _ in range(n):
    uid=input("Enter an employee's UID: ")
    result=is_valid_uid(uid)
    print(result)
```
